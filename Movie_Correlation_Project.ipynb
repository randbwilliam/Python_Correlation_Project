```python
#Import Libraries

import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration to the plots soon to be created
```

```python
#Read in the data
df = pd.read_csv(r'C:\Users\randb\Downloads\movies.csv', encoding = 'latin1')
```

```python
df.head()
```

```python
for col in df.columns:
        percent_missing = np.mean(df[col].isnull())
        print('{} - {}%'.format(col, percent_missing))
```

```python
#Data types for our columns 
df.dtypes
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create correct year of release column by taking the year (the first 4 charactrs in the release column)

df['year_correct'] = df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace = False, ascending = False)
df.head(10)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#Budgethigh correlation
#Company high correlation

#Scatter plot with budget vs gross
plt.scatter(x=df['budget'], y =df['gross'])
plt.title('MatPlotLib: Budget vs Gross Earnings')
plt.xlabel('Gross Earnings (hundred millions)')
plt.ylabel('Budget for Film (hundred milions)')
plt.show
```

```python
#Plot Budget vs Gross using seaborn
plot_2 = sns.regplot(x= 'budget', y= 'gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
plot_2.set(xlabel='Gross Earnings (hundred millions)', ylabel = 'Budget for Film (hundred milions)', 
           title = 'Seaborn: Budget vs Gross Earnings')
```

```python
df.corr(method='pearson')
```

```python
corr_matrix = df.corr(method='pearson')
sns.heatmap(corr_matrix, annot = True)
plt.title('MatPlotLib Heatmap: Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
#Look at company
df_num = df
for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] =  df_num[col_name].cat.codes
df_num.head()
```

```python
corr_matrix_2 = df_num.corr(method='pearson')
sns.heatmap(corr_matrix_2, annot = True)
plt.title('MatPlotLib Heatmap: Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show
```

```python
corr_mat = df_num.corr(method='pearson')
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
